# Sigmoid Function

It will give an output of 0 and 1.

```python
import math

def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
#sigmoid function will give a range between 0 and 1
sigmoid(100)
```

```python
sigmoid(-10) #close to 0
```

```python
sigmoid(1) #close to 1
```

```python
sigmoid(-19)
```

# Tanh Function

It gives an output between -1 and 1.

```python
def tanh(x):
  return (math.exp(x)- math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-1)
```

```python
tanh(57)
```

```python
tanh(-23)
```

```python
tanh(-90)
```

```python
tanh(1)
```

# RELU Function

```python
def relu(x):
  return max(0,x)
```

```python
relu(-7)
```

```python
relu(8)
```

# Leaky Relu

```python
def leaky_relu(x):
  return max(0.1*x, x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
