```python
import numpy as np
from time import time
```

```python
#two list filled with 10000000 random numbers 
n = 10000000
x= np.random.rand(n)
y= np.random.rand(n)
```

# Adding Two List Elementwise

**Non-Vectorized Version**

```python
start_time = time()
z1 = []
for k in range(n):
  z1.append(x[k]+y[k])
end_time = time()
t1= end_time - start_time
print (t1)
```

**Vectorized Version**

```python
start_time = time()
z2 = x+y
end_time = time()
t2= end_time- start_time
print(t2)
```

```python
#verify we got the same answer
(z1 == z2).all()
```

# Matrix Multiplication

**Non-vectorized**

```python
n = 200
A= np.random.rand(n,n)
B= np.random.rand(n,n)
```

```python
#by default the matrix product is all zeros
C1= np.zeros((n,n))
```

```python
start_time = time()
#iterate over all rows and columns
for i in range(n):
  #get ith row of A
  row_a = A[i]
  for j in range(n):
     #get jth column of B
     col_b = B[:, j]

     #compute the i, j entry of C1 as the dot product of the ith row of A with jth column
     entry = 0
     for k in range(n): #dot product
       entry+=row_a[k]*col_b[k]
     C1[i,j]= entry
     
  end_time = time()
  t1= end_time- start_time
  print(t1)
```

# Vectorized 

```python
start_time = time()

#numpy matrix multiplication
C2 = np.dot(A,B)

end_time = time()
t2= end_time- start_time
print(t2)
```

```python
#verify we got the same answer
(C1 == C2).all()
```

```python
np.mean(abs(C1-C2))
#this means the values are close by
```
